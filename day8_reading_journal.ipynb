Exercise 12.2. In this example, ties are broken by comparing words, so words with the same length appear in reverse alphabetical order. For other applications you might want to break ties at ran- dom. Modify this example so that words with the same length appear in random order. Hint: see the random function in the random module.

```python
import random
def sort_by_length_rand(words):
    t = []
    res = []
    for word in words:
       t.append((len(word), random.random(), word))
    t.sort(reverse=True)
    for length, _, word in t:
        res.append(word)
    return res

sort_by_length_rand(["The", "wonderful", "thing", "about", "tiggers"])
```

1. Write a program that reads a word list from a file (see Section 9.1) and prints all the sets of
words that are anagrams.
Here is an example of what the output might look like:

     ['deltas', 'desalt', 'lasted', 'salted', 'slated', 'staled']
     ['retainers', 'ternaries']
     ['generating', 'greatening']
     ['resmelts', 'smelters', 'termless']

Hint: you might want to build a dictionary that maps from a set of letters to a list of words that can be spelled with those letters. The question is, how can you represent the set of letters in a way that can be used as a key?

```python
#sorted list of letters of a word
def original(word):
    letters = list(word) #splits word up into list of letters
    letters.sort()
    letters = ' '.join(letters) #joints the list of letters and inserts space in between
    return letters

original('chicken')

def anagrams(filename):
    dictionary = {}
    for item in open(filename): #takes the words in the file
        spelling = item.strip().lower()
        letters = original(spelling)
        if letters in dictionary:
            dictionary[letters].append(spelling)
        else: dictionary[letters] = [spelling]
    return dictionary


def anagrams_list(dictionary):
    word_list = [] #makes a list
    for words in dictionary.values(): 
        if len(words) > 1:         
            word_list.append(words)
    return word_list 
                      
d = anagrams('/home/sandy/Desktop/words.txt')
anagrams_list(d)

```

2\. Modify the previous program so that it prints the largest set of anagrams first, followed by the second largest set, and so on.

```python
def anagrams_list_sorted(dictionary):
    t = []
    for gram in dictionary.values():
        if len(gram) > 1:
            t.append((len(gram), gram))
    t.sort(reverse=True)
    return t

anagrams_list_sorted(d)


```

```python
def bingo(dictionary):
    highest_bingo = []
    for gram in dictionary.values(): 
        if len(gram[0]) == 8: #checks length of the first word in the anagram
            if len(gram) > len(highest_bingo): #if the number of items in the current anagram are greater than the current number of highest bingo, replace it
                highest_bingo = gram           
    return highest_bingo

bingo(d)
```
